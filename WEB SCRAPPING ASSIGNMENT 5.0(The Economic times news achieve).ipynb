# Importing Libraries

```python
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)
```

```python
date1=[]
author1=[]
vertical1=[]
headline1=[]
description1=[]
```

```python
driver.get('https://economictimes.indiatimes.com/news/politics-and-nation/view-pm-modis-new-cabinet-signals-continuity/articleshow/69605797.cms')
```

```python
# scrapping details 1 june 2019

dt=driver.find_elements_by_xpath("//div[@class='dtc vam artByline']/time")
for i in dt:
    try:
        date1.append(i.text)
    except NoSuchElementException:
        date1.append('-')
        
    
    
at=driver.find_elements_by_xpath("//div[@class='auth eventDone']")
for i in at:
    try:
        author1.append(i.text)
    except NoSuchElementException:
        author1.append('-')
    
    
ve=driver.find_elements_by_xpath("//div[@data-ga-action='Politics']")
for i in ve:
    try:
        vertical1.append(i.text)
    except NoSuchElementException:
        vertical1.append('-')
   
    
    
hd=driver.find_elements_by_xpath("//h1[@class='artTitle font_faus']")
for i in hd:
    try:
        headline1.append(i.text)
    except NoSuchElementException:
        headline1.append('-')
   
    
    
desc=driver.find_elements_by_xpath("//div[@class='artText medium']")
for i in desc:
    try:
        description1.append(i.text)
    except NoSuchElementException:
        description1.append('-')
    
    
```

```python
df1=pd.DataFrame({})
df1['Date']=date1[:1]
df1['Author']=author1[:1]
df1['Vertical']=vertical1[:1]
df1['Healines']=headline1[:1]
df1['Description']=description1[:1]
```
